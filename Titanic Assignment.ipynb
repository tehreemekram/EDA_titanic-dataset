**Question 6**: What is the correlation between different columns?

 **Question 7**: Are there any outliers in the dataset?

```python
array_column= ["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]
for value_column in array_column:
# value_column = 'PassengerId'
  outliers = df[(df[value_column] < df[value_column].mean() - 2 * df[value_column].std()) |
              (df[value_column] > df[value_column].mean() + 2 * df[value_column].std())]
  # print(f"Outliers of column {value_column}:\n",outliers)
  sns.catplot(x="Sex", y=value_column, data=df, kind="box", aspect=1.5)
  plt.title(f"Box Plot of result vs {value_column}")
  plt.show()
```

Yes, there are outliers present in the "Age", "SibSp", "Parch", "Fare" columns. We represent them with the help of box plot for better understanding and visualization.
